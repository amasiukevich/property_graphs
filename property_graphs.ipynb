### Configuring the LLM and Embedding models

```python
from llama_index.llms.ollama import Ollama
from llama_index.embeddings.ollama import OllamaEmbedding
```

```python
llm = Ollama(model="llama3")
embed_model = OllamaEmbedding(model_name="nomic-embed-text")
```

```python
llm.complete("It's a beautiful day!")
```

```python
embed_model.get_text_embedding("It's a beautiful day!")
```

### Loading the data

```python
FILENAME = "data/paul_graham/paul_graham_essay.txt"

from llama_index.core import SimpleDirectoryReader

documents = SimpleDirectoryReader(input_files=[FILENAME]).load_data()
```

### Creating a PropertyGraphIndex

```python
from llama_index.core import PropertyGraphIndex

pgi = PropertyGraphIndex.from_documents(documents=documents, llm=llm, embed_model=embed_model, show_progress=True)
```
